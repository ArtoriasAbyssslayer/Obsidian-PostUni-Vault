```python
import cv2
import os

# Define input
vs = cv2.VideoCapture('video.mp4')

output_file = os.path.join('out.mp4')
frame_width = 500
frame_height = 281
fourcc = cv2.VideoWriter_fourcc(*'FMP4')
video = cv2.VideoWriter(output_file, fourcc,float(25), (frame_width,frame_height))
# Define the tracker
tracker = cv2.TrackerKCF_create()
initialized = False
initBB = tuple([187,113,60,55])
# Tracking Process
while True:
  ret, frame = vs.read()
  if not ret:
    break
  (H,W,D) = frame.shape

  if not initialized:
    tracker.init(frame,initBB)
    initialized = True
  (success, box) = tracker.update(frame)
  if success:
    (x,y,w,h) = [int(v) for v in box]
    output = frame.copy()
    cv2.rectangle(output,(x,y),(x+w,y+h),(0,255,127), 4)
    video.write(output)
vs.release()
video.release()    
```
