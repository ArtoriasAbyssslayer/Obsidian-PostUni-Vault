```python
import cv2 as cv

img = cv.imread("./image.jpg")
from google.colab.patches import cv2_imshow

cv2_imshow(img)
img
```

GET some basic info of the image

```python
(h,w,c) = img.shape
print("width={}, height={},channel={}".format(w,h,c))
```

Image Convertions

```python
# Convert image to grayscale
gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
# Extract RoI

roi = img[120:338,226:538]
cv2_imshow(roi)
```

resize image

```python
resized = cv.resize(roi,(200,200))
cv2_imshow(resized)
(h1,w1,c1) = roi.shape
r1 = 200.0/w1
dim=(200, int(h1*r1))
resized = cv.resize(roi,(0,0),fx=0.5,fy=0.5)
cv2_imshow(resized)
```

```python
# Image rotation
center = (w//2,h//2)
M = cv.getRotationMatrix2D(center,-45,1.0)
rotated = cv.warpAffine(img,M,(w,h))
cv2_imshow(rotated)
```

```python
blurred = cv.GaussianBlur(img,(5,5),0)
cv2_imshow(blurred)
```

```python
output = img.copy()

rectangle = cv.rectangle(output, (226, 120), (538, 338), (0, 0, 255), 2)
cv2_imshow(rectangle)
```

```python
output = img.copy()
cv.putText(output, "Introduction to OpenCV", (10, 30),
    cv.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
# cv2.imshow("Text", output)
# cv2.waitKey(0)
# cv2.imwrite(os.path.join(image_path, 'output_2.jpg'), output)
cv2_imshow(output)
```
